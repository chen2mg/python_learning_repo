# 🔢 Chapter 2: Playing with Words and Numbers
### Not all data is the same!
Robots are very picky about what kind of information you give them. You can't ask a robot to "drive Blue inches" or "set light color to 50".

Today we learn the three main **Data Types**:
1. **String (str)**: Text, words, and messages. Always inside `"quotes"`.
2. **Integer (int)**: Whole numbers. Used for counting things (like 1, 5, 100).
3. **Float**: Decimal numbers. Used for precise math (like 3.14, 10.5).

```python
# Examples of Data Types
message = "Status OK"      
rotations = 5              
wheel_size = 200.5        

print(message)
print(rotations)
print(wheel_size)
```

### 🧮 Robot Math
We can use Python to do math for us! This is useful when we need to calculate how far the robot has traveled.

* `*` means Multiply
* `+` means Add
* `-` means Subtract
* `/` means Divide

```python
# Let's calculate distance!
# If one wheel turn moves the robot 200mm, how far is 3 turns?

one_turn_distance = 200  # millimeters
total_turns = 3

total_distance = one_turn_distance * total_turns

print("Total Distance Traveled (mm):")
print(total_distance)
```
