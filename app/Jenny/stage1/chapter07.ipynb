# ⏳ Chapter 7: While Loops
### Keep Going Until...
A `for` loop counts numbers (1, 2, 3...). But sometimes we don't know how far we need to drive. We just want to drive **UNTIL** we see a wall.

A **While Loop** runs forever as long as a condition is `True`.

### Example: Waiting for Start
Imagine the robot is waiting for a button press to start the match.

```python
start_button_pressed = False
time_seconds = 0

# While the button is NOT pressed, keep waiting
while start_button_pressed == False:
    print("Waiting for start...")
    time_seconds = time_seconds + 1
    
    # Let's pretend the button gets pressed after 3 seconds
    if time_seconds == 3:
        start_button_pressed = True

print("GO! GO! GO!")
```

### 🛠️ Robot Activity: "Search & Rescue"
**Your Mission:**
The robot is looking for an object. It should drive forward until its Distance Sensor sees something close (less than 50mm).

1. Create a variable `distance_to_object` and set it to `1000` (far away).
2. Create a `while` loop that runs as long as `distance_to_object > 50`.
3. Inside the loop, print **"Driving forward... searching..."**.
4. Inside the loop, decrease `distance_to_object` by `200` (Simulating getting closer).
5. After the loop, print **"Object Found! Stopping!"**

```python
distance_to_object = 1000
while distance_to_object > 50:
	print("Driving forward...searching...")
	distance_to_object -= 200
	print("Distance To Object Is",distance_to_object)
print("Object Found!Stopping!")
```

### 🏠 Homework Challenge
**The Battery Drain:**
Set `battery` to `100`. Write a while loop that runs while `battery > 0`.
Inside the loop, print "Robot Running" and subtract 10 from the battery every time.
When the loop finishes, print "Battery Died."

```python
battery = 100
while battery > 0:
    print("Robot Running")
    battery -= 10
print("Battery Died")
```
