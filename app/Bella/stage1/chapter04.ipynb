# 🔀 Chapter 4: The Fork in the Road (If/Else)
### Robots Need to Make Choices
So far, our code just runs from top to bottom. But what if the robot hits a wall? It needs to decide: **"If I hit a wall, stop. Otherwise, keep going."**

We use `if` and `else` statements to create these forks in the road.

### How it Works
Imagine a security guard checking for an ID card.

1. **`if`**: Checks a condition (True/False).
2. **`else`**: What to do if the condition is False (the backup plan).

**Note:** In Python, **indentation (spacing)** is super important! The code inside the `if` statement must be indented (press Tab).

```python
# Example: The Security Guard
has_id_card = False

if has_id_card == True:
    print("Access Granted. Welcome!")
else:
    print("Access Denied. Please leave.")

# Try changing has_id_card to True and run it again!
```
