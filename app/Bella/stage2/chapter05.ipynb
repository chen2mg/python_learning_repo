### Breaking the Loop
Sometimes we want a loop to run, but we need a secret escape hatch.
The command `break` immediately stops the loop and jumps to the code at the very end.

This is perfect for: "Drive forward forever... **UNLESS** you hit a wall."

```python
# Example: Counting to 10, but stopping at 5
for number in range(10):
    print(number)
    if number == 5:
        print("I'm bored. Stopping now.")
        break
```
