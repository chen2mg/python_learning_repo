# 🔄 Chapter 4: The Line Follower (While True)
### The Infinite Loop
Most robot programs don't just run once and stop. They need to keep thinking forever (or until the match ends).

We use `while True:` to create a loop that **never stops** on its own. This is often called the "Main Loop" or "Game Loop."

### The Danger of Infinity
**Warning:** If you run the code below, it will print forever! You have to press the **Stop Button** (the square icon) in Jupyter to kill it.

### 🛠️ Robot Activity: "The Line Tracker"
**Your Mission:**
A Line Follower works by wiggling. It checks the floor color constantly.
1.  Create a variable `light_sensor_value`.
2.  Create a `while True:` loop.
3.  **Inside the loop**:
    * Imagine the sensor reads a value (we will simulate this using `import random` to pretend the robot is moving over lines).
    * If value < 50 (Dark Line): Print "Turn LEFT to find white".
    * Else (White Floor): Print "Turn RIGHT to find line".
4.  **Important:** Add `import time` and `time.sleep(1)` so it doesn't print too fast!

```python
import random
import time

# --- WRITE YOUR MISSION CODE BELOW ---

print("Starting Line Follower...")

while True:
    # 1. Simulate Sensor (Pick random number 0-100)
    light_sensor_value = random.randint(0, 100)
    print("Sensor reads:", light_sensor_value)
    
    # 2. Write your Logic (If Dark -> Left, Else -> Right)
    if light_sensor_value<50:
        print("Turn left to find white.")
    else: print("Turn right to find line")


    
    # 3. Sleep for 1 second so we can read the text
    time.sleep(5)
```

### 🏠 Homework Challenge
**The Emergency Kill Switch:**
Inside the `while True` loop, add a check:
`if button_pressed == True:`
Then use the command `break` to stop the loop.

```python
import random
import time



print("Starting Line Follower...")

while True:
    light_sensor_value = random.randint(0, 100)
    print("Sensor reads:", light_sensor_value)
    if_button_pressed = False

    if light_sensor_value<50:
        print("Turn left to find white.")
    else: print("Turn right to find line")

    if if_button_pressed == True:
        break
    else: print("button was not pressed")


    
    # 3. Sleep for 1 second so we can read the text
    time.sleep(3)
```
