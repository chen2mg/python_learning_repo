### Don't Repeat Yourself!
Imagine you want your robot to drive in a square. You could write:
* `Drive Forward`
* `Turn Right`
* `Drive Forward`
* `Turn Right`...

That takes forever to type! Instead, we use a **For Loop** to tell the robot: **"Do this 4 times."**

### The Range Command
We use `range()` to tell the computer how many times to count.

**Try running this code:**

```python
# Repeat 3 times
print("Starting Loop...")

for number in range(3):
    print("Beep!")

print("Loop Finished!")
```
