### Breaking the Loop
Sometimes we want a loop to run, but we need a secret escape hatch.
The command `break` immediately stops the loop and jumps to the code at the very end.

This is perfect for: "Drive forward forever... **UNLESS** you hit a wall."

```python
# Example: Counting to 10, but stopping at 5
for number in range(10):
    print(number)
    if number == 5:
        print("I'm bored. Stopping now.")
        break
```

```python
import random

# --- WRITE YOUR MISSION CODE BELOW ---
fuel = 5
while True:
   # 1. Check Fuel (if 0, break)
    if fuel == 0:
        print("Out of Gas!")
        break
    else:
        print("Driving...")
        fuel = fuel - 1
    # 2. Check Wall (if random > 8, break)
    if random.randint(1, 10) > 8:
        print("Hit a Wall!")
        break

    # 3. Drive logic
```
