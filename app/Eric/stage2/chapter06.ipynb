# 📈 Chapter 6: Smooth Operator (Ramping)
### Using the Loop Variable
When we write `for i in range(10):`, the variable `i` counts up: 0, 1, 2, 3...

We can use this number to change the robot's speed gradually.
This is called **Ramping**. It stops the robot from doing a wheelie by accelerating too fast!

```python
# Example: Watch the number 'i' grow
for i in range(5):
    print("Speed is now:", i)
```
