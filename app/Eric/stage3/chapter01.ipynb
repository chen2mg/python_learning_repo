# 🤖 Chapter 1: The Blueprint and the Robot

Welcome! Today we will learn something new and exciting: **Classes and Objects**.

Think about this:
- A **class** is like a *blueprint* 🏗️
- An **object** is the *real robot* built from that blueprint 🤖

Let's learn step by step!

## 🧠 What is a Class?

A **class** is a plan or template.

Example:
- A robot factory has a blueprint
- That blueprint describes what every robot should look like

But the blueprint itself is NOT a robot!

Let's create our first class 👇

```python
a =8
```

```python
a
```

```python
# This is a class (a blueprint)
class Robot:
    pass  # pass means "nothing here yet"
```

## 🤖 What is an Object?

An **object** is a real robot made from the class.

We can create (build) a robot using the class!

```python
# Create a robot (object)
Huey = Robot()

print("I built a robot!", Huey)
```

```python
# Another example
class Dog:
    pass

macy = Dog()
mina = Dog()

print("I have two dogs!", macy, mina)
```

👉 You just created your first robot object!

We can create MANY robots from the same class:

```python
class human:
    def __init__(self,name):
        self.name = name

eric = human('e')
grace = human('g')


print(eric.name)
```

```python
robot1 = Robot()
robot2 = Robot()
robot3 = Robot()

print("Robot 1:", robot1)
print("Robot 2:", robot2)
print("Robot 3:", robot3)
```

## 🎯 Key Idea

- **Class = Blueprint 🏗️**
- **Object = Real Robot 🤖**

One class → many robots!
