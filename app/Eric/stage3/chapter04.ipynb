# 🤖 Chapter 4: Setting Up Your Robot (Calibration)

Now our robot can THINK and ACT...

But before a real robot starts, it must **prepare itself**!

This is called **Calibration** ⚙️

👉 Calibration makes sure:
- Sensors are correct
- Motors start in the right position
- Robot is ready to move safely


## 🧠 What is `__init__`?

`__init__` is a special method that runs **automatically** when we create a robot.

👉 Think of it like:
- "When robot turns ON, do these things first"

Perfect place for calibration!

## ⚙️ Simulated VEX Devices

We will simulate VEX hardware so we can learn easily.

```python
class Motor:
    def __init__(self, port, reversed=False):
        self.port = port
        self.reversed = reversed

    def spin(self, direction):
        print(f"Motor on {self.port} spinning {direction}")

    def stop(self):
        print(f"Motor on {self.port} stopped")


class MotorGroup:
    def __init__(self, m1, m2):
        self.m1 = m1
        self.m2 = m2

    def spin(self, direction):
        self.m1.spin(direction)
        self.m2.spin(direction)

    def stop(self):
        self.m1.stop()
        self.m2.stop()


class Inertial:
    def set_heading(self, value):
        print(f"Gyro heading set to {value}")


class Bumper:
    def __init__(self):
        self._pressed = False

    def pressing(self):
        return self._pressed

    def simulate_press(self):
        self._pressed = True


m1 = MotorGroup()
```

## 🤖 Build Robot with Calibration

Now we create a robot that:
- Sets gyro to 0
- Moves arm until bumper is pressed
- Stops safely


```python
class family:
    def __init__(self,people):
        self.people = people

class human:
    def __init__(self, name = "Ming"): # by default, it is Ming
        self.name = name # self shared varibale in this class
        another_name = "abc" # this variable is not shared

    def print_name(self):
        print(self.name)
        #print(another_name) # this will give error message, since another_name is not defined

h1 = human("GRACE")

h1.print_name()



# an object of a class can be attribute of another class
# nested class
```

```python
class Robot:
    def __init__(self):
        print("Robot is starting... 🔋")

        # Setup devices
        self.inertial = Inertial()
        self.bumper = Bumper()

        motor_a = Motor("PORT6")
        motor_b = Motor("PORT12")
        self.BeamArm = MotorGroup(motor_a, motor_b)

        # Run calibration automatically
        self.calibrate()

    def calibrate(self):
        print("Starting calibration... ⚙️")

        # Step 1: Reset gyro
        self.inertial.set_heading(0)

        # Step 2: Move arm until bumper pressed
        print("Moving arm to home position...")

        steps = 0  # prevent infinite loop in simulation

        while not self.bumper.pressing():
            self.BeamArm.spin("forward")
            steps += 1

            # simulate bumper press after few steps
            if steps > 3:
                self.bumper.simulate_press()

        # Stop arm
        self.BeamArm.stop()

        print("Calibration complete ✅")


# Create robot (calibration runs automatically!)
robot = Robot()
```

👉 Notice:
- `__init__` runs automatically
- `calibrate()` is called inside it
- Robot prepares itself before doing anything else


# 🎯 Key Ideas

- `__init__` runs when object is created
- Use it to setup robot
- Calibration ensures safe and correct behavior
- Real robots ALWAYS calibrate first
