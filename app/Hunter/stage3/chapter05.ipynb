## 🧠 What is an Autonomous Routine?

An **autonomous routine** is a list of steps the robot follows by itself.

Example:
1. Drive forward
2. Turn
3. Grab object
4. Move somewhere else

👉 Just like giving your robot a **plan**!

## 🤖 Example: Real Competition Routine

This is a simplified version of a real VEX competition task.

```python
class Robot:
    def __init__(self):
        pass

    def move_forward(self, distance):
        print(f"Driving forward {distance} cm 🚗")

    def turn(self, angle):
        print(f"Turning {angle} degrees 🔄")

    def grab(self):
        print("Grabbing object 🤏")

    def release(self):
        print("Releasing object ✋")

    def perform_task_mix_and_match(self):
        print("Starting mission... 🚀")

        # Step 1: Go to object
        self.move_forward(50)
        self.turn(90)

        # Step 2: Grab it
        self.grab()

        # Step 3: Move to target
        self.turn(-90)
        self.move_forward(30)

        # Step 4: Release
        self.release()

        print("Mission complete ✅")


robot = Robot()
robot.perform_task_mix_and_match()
```

## 🔍 Let's Understand the Code

👉 What is happening?

- The robot follows steps in order
- Each step calls a method
- This creates a full mission!

💡 This is exactly how real VEX robots work!
