# 🔑 Chapter 2: The Two-Key Turn (AND / OR)
### Double Checking
Sometimes, checking one thing isn't enough. Imagine a nuclear submarine. To launch a missile, **two** people have to turn their keys at the exact same time.

We call these **Logical Operators**:
1.  **`and`**: BOTH things must be True.
2.  **`or`**: EITHER thing can be True.

### The "AND" Check (Strict)
The robot will only move if it has battery **AND** the path is clear.

```python
battery = 50
path_is_clear = True

# Try changing one of these to False and see what happens!
if battery > 0 and path_is_clear == True:
    print("Systems Go! Moving Forward.")
else:
    print("Cannot move. Check battery or path.")
```

### The "OR" Check (Flexible)
The robot should stop if it hits a wall **OR** if the emergency button is pressed.

```python
hit_wall = False
emergency_button = True

if hit_wall == True or emergency_button == True:
    print("STOPPING ROBOT IMMEDIATELY!")
else:
    print("Driving...")
```
