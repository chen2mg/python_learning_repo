## 🧠 What is an Autonomous Routine?

An **autonomous routine** is a list of steps the robot follows by itself.

Example:
1. Drive forward
2. Turn
3. Grab object
4. Move somewhere else

👉 Just like giving your robot a **plan**!

## 🤖 Example: Real Competition Routine

This is a simplified version of a real VEX competition task.

```python
class Robot:
    def __init__(self):
        pass

    def move_forward(self, distance):
        print(f"Driving forward {distance} cm 🚗")

    def turn(self, angle):
        print(f"Turning {angle} degrees 🔄")

    def grab(self):
        print("Grabbing object 🤏")

    def release(self):
        print("Releasing object ✋")

    def perform_task_mix_and_match(self):
        print("Starting mission... 🚀")

        # Step 1: Go to object
        self.move_forward(50)
        self.turn(90)

        # Step 2: Grab it
        self.grab()

        # Step 3: Move to target
        self.turn(-90)
        self.move_forward(30)

        # Step 4: Release
        self.release()

        print("Mission complete ✅")


robot = Robot()
robot.perform_task_mix_and_match()
```

## 🔍 Let's Understand the Code

👉 What is happening?

- The robot follows steps in order
- Each step calls a method
- This creates a full mission!

💡 This is exactly how real VEX robots work!

```python
#COach Chen I made a random router it does not perfectally works and i am still fixing it
import random
import time

class Robot:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.angle = 0
        self.visited = [(0, 0)]

    def move_forward(self, distance):
        print(f"Driving forward {distance} cm 🚗")
        self.x += distance

    def turn(self, angle):
        print(f"Turning {angle} degrees 🔄")
        self.angle += angle

    def grab(self):
        print("Grabbing object 🤏")

    def release(self):
        print("Releaesing object")

    def random_mission(self):
        print("=" * 50)
        print("RANDOM ROUTER STARTED")
        print("=" * 50)

        directions = [90, -90, 45, -45, 180, 0]
        distances = [100, 200, 150, 300, 50]
        
        print(f"\nStart position: ({self.x}, {self.y})")

        for i in range(5):
            turn = random.choice(directions)
            dist = random.choice(distances)
            
            print(f"\nStep {i+1}:")
            self.turn(turn)
            time.sleep(0.3)
            self.move_forward(dist)
            time.sleep(0.3)
            
            if random.random() > 0.5:
                self.grab()
                time.sleep(0.3)
                self.release()
                time.sleep(0.3)
            
            self.visited.append((self.x, self.y))
            print(f"Current: ({self.x}, {self.y})")

        print(f"\nFinal position: ({self.x}, {self.y})")
        print(f"Steps taken: {len(self.visited)}")

        print("\n" + "=" * 50)
        print("CALCULATING RETURN ROUTE")
        print("=" * 50)

        return_steps = []
        for i in range(len(self.visited) - 1, 0, -1):
            prev = self.visited[i - 1]
            curr = self.visited[i]
            
            dx = prev[0] - curr[0]
            dy = prev[1] - curr[1]
            
            if dx != 0:
                dist = abs(dx)
                direction = 180 if dx > 0 else 0
                return_steps.append(("forward", dist, direction))
            
            if dy != 0:
                dist = abs(dy)
                direction = 90 if dy > 0 else -90
                return_steps.append(("forward", dist, direction))

        print(f"\nReturning home...\n")
        for step in return_steps:
            action, dist, turn = step
            if turn != 0:
                self.turn(turn)
                time.sleep(0.3)
            self.move_forward(dist)
            time.sleep(0.3)

        print(f"\nBack at start: ({self.x}, {self.y})")
        print("=" * 50)
        print("MISSION COMPLETE")
        print("=" * 50)

robot = Robot()
robot.random_mission()
```
