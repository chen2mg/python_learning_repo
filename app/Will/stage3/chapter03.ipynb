# 🤖 Chapter 3: Robot Actions (Behavior)

So far, our robots can **remember things** (attributes)...

But they can't DO anything yet 😴

Today we will fix that!

We will learn about **Methods (Functions inside a class)**.

👉 Methods tell the robot what actions to perform!

## 🧠 What is a Method?

A **method** is a function inside a class.

Think of it like:
- Attributes = what the robot *has*
- Methods = what the robot *does*

Examples:
- drive forward 🚗
- detect distance 📏
- check bumper 🔘

## ⚙️ Let's Build a Robot with Actions


```python
class Robot:
    def __init__(self, name):
        self.name = name

    # Method: drive forward
    def drive_forward(self):
        print(self.name, "is driving forward 🚗")

    # Method: detect distance
    def detect_front_distance(self):
        print(self.name, "is checking distance 📏")
        return 100  # pretend distance is 100 mm

    # Method: check bumper
    def is_bumper_pressed(self):
        print(self.name, "is checking bumper 🔘")
        return False

# Create robot
robot = Robot("Huey")

# Call methods
robot.drive_forward()
distance = robot.detect_front_distance()
print("Distance:", distance)

pressed = robot.is_bumper_pressed()
print("Bumper pressed:", pressed)
```

👉 Notice:
- We use `robot.method_name()` to call a method
- Methods can **return values** (like distance)


## 🎮 Make Decisions (Robot Logic)

Now let's make the robot **think**!

If something is in front → STOP
Otherwise → MOVE

```python
distance = robot.detect_front_distance()

if distance < 50:
    print("Too close! Stop! 🛑")
else:
    robot.drive_forward()
```

👉 This is how real robots make decisions!


## 🔌 VEX-Style Robot Example

Now let's make it look more like a real VEX robot!

```python
# Simulated VEX devices
class Distance:
    def object_distance(self):
        return 80  # pretend value

class Bumper:
    def pressing(self):
        return False

class Robot:
    def __init__(self, name):
        self.name = name
        self.distance = Distance()
        self.bumper = Bumper()

    def drive_forward(self):
        print(self.name, "drives forward 🚗")

    def check_and_move(self):
        if self.distance.object_distance() < 50:
            print("Object detected! Stop! 🛑")
        elif self.bumper.pressing():
            print("Bumper hit! Stop! 🔘")
        else:
            self.drive_forward()

# Test robot
robot = Robot("VEXBot")
robot.check_and_move()
```

# 🎯 Key Ideas

- Methods = actions
- Call methods using `robot.method()`
- Methods can return values
- Robots use methods + logic to make decisions
