# 🔄 Chapter 6: Doing it Again (For Loops)
### Don't Repeat Yourself!
Imagine you want your robot to drive in a square. You could write:
* `Drive Forward`
* `Turn Right`
* `Drive Forward`
* `Turn Right`...

That takes forever to type! Instead, we use a **For Loop** to tell the robot: **"Do this 4 times."**

### The Range Command
We use `range()` to tell the computer how many times to count.

**Try running this code:**

```python
# Repeat 3 times
print("Starting Loop...")

# range(4) means from 0 to 3, such as 0, 1, 2, 3
# range(5) means from 0 to 4,such as 0, 1, 2, 3, 4
for number in range(4):
    print("Beep!", number)

print("Loop Finished!")
```
