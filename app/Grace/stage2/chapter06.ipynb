# 📈 Chapter 6: Smooth Operator (Ramping)
### Using the Loop Variable
When we write `for i in range(10):`, the variable `i` counts up: 0, 1, 2, 3...

We can use this number to change the robot's speed gradually.
This is called **Ramping**. It stops the robot from doing a wheelie by accelerating too fast!

```python
# Example: Watch the number 'i' grow
for i in range(5):
    print("Speed is now:", i)
```

### 🛠️ Robot Activity: "Soft Start Acceleration"
**Your Mission:**
The robot needs to go from Speed 0 to Speed 100 smoothly over 10 steps.

1.  Write a `for` loop that runs 10 times (`range(10)`).
2.  Inside the loop, create a variable `current_speed`.
    * Math Tip: If `i` goes 0-9, multiply it by 10 to get 0-90.
    * Formula: `current_speed = i * 10`.
3.  Print "Accelerating... Setting Motor to speed: " and the number.

```python
# --- WRITE YOUR MISSION CODE BELOW ---

# 1. Create the Loop
for i in range(10):
    
    # 2. Calculate Speed (i * 10)
    current_speed = i * 10 

    # 3. Print Command
    print("Accelerating... Setting Motor to speed", current_speed)
```
