### Reading a Map
In Phase 1, we learned that a **List** is like a backpack holding many items.
We can use a `for` loop to look at every single item in the backpack, one by one.

This is perfect for robots. We can give the robot a list of instructions (a "Script") and it will follow them in order.

```python
# Example: The Shopping List
grocery_list = ["Apples", "Bananas", "Milk", "Cookies"]

# The loop runs 4 times (once for each item)
for item in grocery_list:
    print("I need to buy:", item)
```
