Real autonomous robots use something called a **State Machine**. It means the robot is always in one "Mode" (or State).
* **State 1:** "Searching" (Looking for a cube)
* **State 2:** "Driving" (Going to the cube)
* **State 3:** "Scoring" (Dropping the cube)

### 🛠️ The Final Project: "The Autonomous Champion"
**Your Mission:**
Write a full brain for your robot.

1.  Create a variable `robot_state` and set it to `"Searching"`.
2.  Create a `while True:` loop.
3.  Inside, check the states:
    * `if robot_state == "Searching":`
        * Print "Spinning... Looking for object..."
        * **Change State:** `robot_state = "Driving"` (We found it!)
    * `elif robot_state == "Driving":`
        * Print "Driving forward..."
        * **Change State:** `robot_state = "Scoring"` (We arrived!)
    * `elif robot_state == "Scoring":`
        * Print "Arm Up! Dropping Cube!"
        * **Change State:** `robot_state = "Finished"`
    * `elif robot_state == "Finished":`
        * Print "Mission Complete."
        * **break** (End the program)
    
4.  Add `import time` and `time.sleep(2)` so you can watch the brain switch modes.

```python
import time

# --- WRITE YOUR FINAL PROJECT CODE BELOW ---

# 1. Initial State
robot_state = "Searching"

# 2. The Game Loop
while True:
    # 3. The Logic Chain (If / Elif / Elif)
    if robot_state == "Searching":
        print("Spinning... Looking for object.")
        robot_state = "Driving"
    elif robot_state == "Driving":
        print("Driving forward...")
        robot_state = "Scoring"
    elif robot_state == "Scoring":
        print("Arm Up! Dropping cube")
        robot_state = "Finished"
    elif robot_state == "Finished":
        print("Mission Complete.")
        break

    # 4. Sleep to slow it down
    time.sleep(2)
```
