### Using Other People's Code
Python is cool because you don't have to write everything yourself. You can "import" tools that other people built.

Think of a **Library** (or Module) like a DLC pack for a video game. It adds new powers to your robot.

### Rolling the Dice
To use a library, we have to tell Python to grab it first using `import`.

**Try running this code multiple times to see different results:**

```python
import random

# Pick a random number between 1 and 100
random_speed = random.randint(1, 100)

print("Robot Speed set to:")
print(random_speed)
```
