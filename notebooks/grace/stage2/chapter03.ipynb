### Wheels within Wheels
Sometimes, asking one question leads to another question.
* **Question 1:** Do I see a ball? (If No, keep driving. If Yes, go to Q2).
* **Question 2:** Is the ball Red or Blue?

This is called **Nesting**. It means putting an `if` statement *inside* another `if` statement (like a Russian nesting doll).

```python
see_ball = True
ball_color = "Blue"

if see_ball == True:
    print("I see a ball! Investigating...")
    
    # This check ONLY happens if see_ball is True
    if ball_color == "Red":
        print("It is the enemy! Attack!")
    else:
        print("It is friendly. Collect it.")
        
else:
    print("Scanning area... nothing found.")
```
