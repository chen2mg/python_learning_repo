### Giving Answers Back
Functions are like employees. Sometimes you just tell them to do something (like `print`).
But sometimes, you ask them a question, and you need them to **give you an answer**.

We use the keyword `return` to send an answer back to the main program.

```python
# Example: A math machine
def double_number(number):
    answer = number * 2
    return answer  # Send the answer back!

# We save the answer in a variable
result = double_number(10)
print("The result is:", result)
```
