# 🤖 Chapter 2: Robot Specs (State)

Today we will learn how robots **remember things about themselves**!

In programming, we call this **Attributes (Member Variables)**.

Think about a real robot:
- It has a name 🏷️
- It has sensors 👀
- It has parts like wheels ⚙️

All of these are **properties** of the robot.

## 🧠 What are Attributes?

Attributes are **things a robot knows about itself**.

Example:
- name = "Robo1"
- has a distance sensor
- has a bumper sensor

We store these inside a class using something called **self**.

## 🔑 What is `self`?

`self` means:

👉 "THIS robot"

Every robot has its **own** data.

So `self.name` means:
- "this robot's name"

Let's see it in action 👇

```python
# Define a robot class with attributes
class Robot:
    def __init__(self, name):
        self.name = name  # each robot remembers its own name

# Create two robots
robot1 = Robot("Huey")
robot2 = Robot("Scorpion")

print(robot1.name)
print(robot2.name)
```

👉 Even though both robots come from the SAME class,
they have DIFFERENT names!

That’s the power of `self` 🎉

Now let's give our robot more parts!

We will add:
- brain

```python
# Simulated VEX devices (just placeholders for now)
class Brain:
    pass

class Distance:
    pass

class Bumper:
    pass

class Optical:
    pass

# Robot with more attributes
class Robot:
    def __init__(self, name):
        self.name = name
        self.brain = Brain()
        self.distance_sensor = Distance()
        self.bumper = Bumper()
        self.optical = Optical()

# Create a robot
robot = Robot("Huey")

print("Robot name:", robot.name)
print("Has brain:", robot.brain)
print("Has distance sensor:", robot.distance_sensor)
```

# 🎯 Key Ideas

- Attributes = robot's properties
- `self` = "this robot"

- `self.name`
