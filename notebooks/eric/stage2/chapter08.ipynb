### One Function, Many Uses
In Phase 1, we wrote simple functions like `drive_forward()`.
But what if we want to drive forward **Fast** sometimes, and **Slow** other times?

We don't want to write two different functions. We want ONE function that accepts options.
We call these options **Parameters** (or Arguments).

```python
# Example: A function that takes a Name as an input
def say_hello(name):
    print("Hello Commander", name)

# Now we can use it for anyone!
say_hello("Iron Giant")
say_hello("Wall-E")
```
