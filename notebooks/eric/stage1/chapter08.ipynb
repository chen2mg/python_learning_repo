# 📦 Chapter 8: Coding Shortcuts (Functions)
### Creating Your Own Blocks
In VEX blocks, you can make "My Blocks." In Python, we call these **Functions**.

A function is a way to group a bunch of code together and give it a name. This way, you don't have to write the same code over and over again.

### How to Define a Function
We use the keyword `def` (short for Define).

**Example:**

```python
# 1. DEFINE the function (This just memorizes the steps, it doesn't run them yet)
def cheer_for_team():
    print("Go Team!")
    print("You can do it!")
    print("VEX Robotics is cool!")

# 2. CALL the function (This actually runs the code)
cheer_for_team()
```
