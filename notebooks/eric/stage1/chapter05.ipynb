### The Robot's To-Do List
Sometimes we want to give the robot a whole list of instructions at once, like a treasure map.

In Python, a **List** holds many items in one variable. We use square brackets `[ ]` to make a list.

### Picking Items from a List
Imagine a shelf of boxes. The computer counts starting at **ZERO**, not one!

* `[0]` is the first item.
* `[1]` is the second item.
* `[2]` is the third item.

```python
# A list of robot parts
parts = ["Motor", "Sensor", "Wheel", "Gear"]

# Let's grab the first part (Index 0)
print("The first part is:")
print(parts[0])

# Let's grab the third part (Index 2)
print("The third part is:")
print(parts[2])
```
