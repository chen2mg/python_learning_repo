# 🚥 Chapter 1: The Traffic Light (Elif)
### It's Not Just Black and White
In Phase 1, we learned `if` and `else`. That works great for "Yes" or "No" questions.

But what if we have **many** options? Like a traffic light:
* **Red** = Stop
* **Green** = Go
* **Yellow** = Slow Down

We use the command `elif` (which stands for "Else If") to check multiple options in a row.

### How ELIF Works
The computer checks them one by one, from top to bottom. As soon as it finds one that is `True`, it runs that code and **skips the rest**.

```python
traffic_light_color = "Green"

if traffic_light_color == "Red":
    print("STOP! Danger!")
elif traffic_light_color == "Yellow":
    print("Slow down...")
elif traffic_light_color == "Green":
    print("GO! GO! GO!")
else:
    print("Light is broken. Be careful!")
```
