# ❓ Chapter 3: Asking Questions
### True or False?
Robots make decisions using "Boolean Logic". This sounds fancy, but it just means the robot only understands **Yes (True)** or **No (False)**.

We ask the robot questions using **Comparison Operators**:
* `==` : Is it equal to? (Notice the double equals sign!)
* `>` : Is it greater than?
* `<` : Is it less than?

### Checking the Battery
Let's ask the computer if our battery is full.

```python
battery_level = 100

# Question: Is the battery equal to 100?
is_full = (battery_level == 100)

print("Is the battery full?")
print(is_full)
```

### Checking the Sensors
Imagine our VEX IQ Lift sensor tells us how high the arm is.
* If the height is **less than (<) 500**, the arm is safe.
* If the height is **greater than (>) 500**, it might hit the roof!

```python
lift_height = 650
max_safe_height = 500

# Question: Is the lift too high?
is_too_high = (lift_height > max_safe_height)

print("Is the arm too high?")
print(is_too_high)
```
